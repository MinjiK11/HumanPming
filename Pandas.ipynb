### Aggregate Statistics (Groupby)

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pokemon_data.csv')

df

df['count']=1

df.groupby(['Type 1','Type 2']).count()['count']
```

### Working with large amounts of data

```python
new_df=pd.DataFrame(columns=df.columns)

for df in pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pokemon_data.csv', chunksize=5): # chunksize: 몇개씩 묶을건지
  results=df.groupby(['Type 1']).count()

  new_df=pd.concat([new_df, results])
```
